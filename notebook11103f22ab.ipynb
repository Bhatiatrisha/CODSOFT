```python
test_data=pd.read_csv("/kaggle/input/test-file/tested.csv")
test_data.head()
```

```python
Women=test_data.loc[test_data.Sex=='female']["Survived"]
rate_women=sum(Women)/len(Women)
print("% of women who survived: ", rate_women*100)
```

```python
men= test_data.loc[test_data.Sex=='male']["Survived"]
rate_men=sum(men)/len(men)
print("% of men who survived: ", rate_men*100)
```

```python
fare=test_data.loc[test_data.Fare>7.0000]["Survived"]
rate_fare=sum(fare)/len(fare)
print("% of people with fare more than 7 who survived: ", rate_fare*100)
print(len(fare))
```

```python
fare=test_data.loc[test_data.Fare<=7.0000]["Survived"]
rate_fare=sum(fare)/len(fare)
print("% of people with fare less than 7 who survived: ", rate_fare*100)
print(len(fare))
```

```python
max_fare=max(test_data.Fare)
print("max_fare: ", max_fare)
min_fare=min(test_data.Fare)
print("min_fare: ", min_fare)
```

```python
test_data=test_data.dropna(subset=['Sex','Fare','Pclass'])
print(len(test_data.Fare))
```

```python
from sklearn.ensemble import RandomForestClassifier
y= test_data["Survived"]
features=["Sex","Fare","Pclass"]
X= pd.get_dummies(test_data[features])
model = RandomForestClassifier(n_estimators=100, max_depth=5, random_state=1)
model.fit(X, y)
```

```python
data=pd.read_csv("/kaggle/input/titanic/train.csv")
data.head()
```

```python
X_test = pd.get_dummies(data[features])
prediction=model.predict(X_test)
output = pd.DataFrame({'PassengerId': data.PassengerId, 'Survived': prediction})
print(output)
```

```python
print(data["Survived"])
```

```python
num_correct = 0
for i in range(len(data["Survived"])):
    if data.Survived[i] == prediction[i]:
        num_correct += 1
```

```python
print(num_correct/len(data.Survived)*100)
```
